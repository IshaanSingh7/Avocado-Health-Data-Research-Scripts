```python
hosp_df = pd.read_csv('Hospital_General_Information.csv')
```

```python
hosp_df.head()
```

```python
birthing_df = hosp_df[hosp_df['Meets criteria for birthing friendly designation'] == 'Y']
```

```python
birthing_df.columns
```

```python
import pandas as pd

# Example: converting birthing_df to the new structure
new_df = pd.DataFrame()

# Mapping existing fields
new_df['name'] = birthing_df['Facility Name']
new_df['phone'] = birthing_df['Telephone Number']
new_df['state'] = birthing_df['State']
new_df['county'] = birthing_df['County/Parish']
new_df['address'] = birthing_df['Address'] + ', ' + birthing_df['City/Town']  # combine street + city
new_df['zip code'] = birthing_df['ZIP Code']

# Fields not in the dataset — placeholder empty strings
new_df['website'] = ''
new_df['Contact Email'] = ''

# Construct a "Description & Services offered" field using keywords from existing columns
def build_description(row):
    services = []
    if row.get('Hospital Type'):
        services.append(f"Type: {row['Hospital Type']}")
    if row.get('Hospital Ownership'):
        services.append(f"Ownership: {row['Hospital Ownership']}")
    if row.get('Emergency Services') == 'Yes':
        services.append("Emergency Services Available")
    if row.get('Meets criteria for birthing friendly designation') == 'Yes':
        services.append("Birthing Friendly Designation")
    # Ratings summary
    if row.get('Hospital overall rating'):
        services.append(f"Overall Rating: {row['Hospital overall rating']}")
    return '; '.join(services)

new_df['Description & Services offered'] = birthing_df.apply(build_description, axis=1)

# Optional: save to CSV
new_df.to_csv('avo_birthing_centers.csv', index=False)
```
